```python
#instalar librerias
#!pip install scipy matplotlib
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

#definir funcion
def rosenbrock(x):
  return (1 - x[0])**2 + 100* (x[1] - x[0]**2)**2

#optimizar funcion
x0 = np.array([-2, 2]) #punto inicial de optimización
result = optimize.minimize(rosenbrock, x0, method='BFGS')

#graficar función
#crear una ciadricula de puntos
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x,y)
Z = rosenbrock([X, Y])

#trazar las lineas de contorno de la función
plt.contour(X, Y, Z, levels=50, cmap='jet')

#trazar el mínimo encontrado
plt.plot(result.x[0], result.x[1], 'ro')

#Etiquetas y títulos del gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('función de Rosenbrock')

#Mostrar el gráfico
plt.show()
```
